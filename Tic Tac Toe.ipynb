```python
from IPython.display import clear_output
import random

def choose_markers(): #works
    '''
    OUTPUT: (player 1 marker, player 2 marker)
    '''
    player1 = 'novalue'
    player2 = 'O'
    while player1 != 'X' and player1 != 'O':
        player1 = input("Player 1, pick a marker X or O: ").upper()
    if player1 == player2:
            player2 = 'X'
    
    return (player1, player2)
        
def reset_board():
    #note: game will ignore the 0 index
    board = [' '] * 10 
    return board
    
def display_board(board):
    i = 1
    while i <= 9:
        print(" {} | {} | {} ".format(board[i], board[i+1], board[i+2]))
        i += 3
        
def input_check(): 
    while True:
        position = int(input("Choose a position (number): "))
        if board[position] != ' ' or position == 0 or position > 9:
            print("Wrong input, position is already taken or invalid. Choose another one.")
        else:
            return position
    
def place_marker(board, marker, position):
    clear_output()
    board[position] = marker
    display_board(board)

def win_check(board, marker): #works
    i = 1
    while i <=9:
        if board[i] == board[i+1] == board[i+2] == marker:  #rows
            return True
        i += 3
    for i in range(1,3):
        if board[i] == board[i+3] == board[i+6] == marker:  #column
            return True
    if board[1] == board[5] == board[9] == marker:  #diagonal 1
        return True
    if board[3] == board[5] == board[7] == marker:  #diagonal 2
        return True
    return False
        
    
def replay(play): #works
    decision = input("Do you want to replay? Write 'y' for Yes and 'n' for No.\n").lower()
    if decision == 'y':
        play = True
        clear_output()
        return play
    elif decision == 'n':
        print('Thanks for playing Tic Tac Toe!')
        play = False
        return play
    else:
        print('Wrong input')
        replay(play)
    
play = True
while play:
    player1, player2 = choose_markers()
    board = reset_board()
    display_board(board)
    start = random.randint(1, 2)  #either player 1 or 2 will start
    stop = start + 9
    
    for i in range(start,stop):
        if i%2 != 0: #p1 moves
            print("Player 1 move.\n")
            position = input_check()
            place_marker(board, player1, position)
            if win_check(board, player1):
                print("Player 1 wins!")
                break
        else: #p2 moves
            print("Player 2 move. \n")
            position = input_check()
            place_marker(board, player2, position)
            if win_check(board, player2):
                print("Player 2 wins!")
                break
    
    play = replay(play)
```
